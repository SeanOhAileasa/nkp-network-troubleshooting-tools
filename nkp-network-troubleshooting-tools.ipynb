###### Hardware Tools
***

After completing this topic, you should be able to: ``use hardware tools to fix network hardware issues``.

All right, in this presentation, we'll take a look at some of the physical tools that you will likely encounter in a network troubleshooting environment. 

< [Table of Contents](#top) | [References](#references) >
<a id="crimper"></a>
***
###### Crimper
***

Tool used to connect an actual RJ45 plug...

... or possibly an RJ11, to the copper wire itself (connect twisted pair cabling into the RJ45 plug).

Its often used to strip back the plastic shielding and expose the copper wire itself - crimping tools often strip and cut twisted pair cabling.

Then, you quite literally can use this to clamp down a new connector to the exposed wiring, so if a cable breaks, for example, right at the connector, you can just strip back some of the wire and crimp a new connection on. 

< [Table of Contents](#top) | [References](#references) >
<a id="CableTester"></a>
***
###### Cable Tester
***

Used to test whether the cable is functioning - tests pin configurations - commonly used for voice and data cables and you can identify and troubleshoot faulty cables. 

Some of the common tests include determining the length and return loss.

> Is it's too long?

> What kind of return loss you are experiencing on that cable?

Can get a ``Wiremap`` so that you know that the connections on one end match the connections on the other end. 

So you quite literally just plug the cable into the tester, both ends, and it tells you what the wiring configuration is.

Can also just use that to verify whether or not it's a straight-through cable, a crossover cable, or a rollover cable. 

< [Table of Contents](#top) | [References](#references) >
<a id="PunchDownTool"></a>
***
###### Punch Down Tool
***

A punch down tool is used to connect twisted pair cable into a 66 or a 110 block and they are quite simply used to connect those wires into the actual punch down block (or a patch panel most commonly these days).

In essence, don't have an actual RJ45 connection going into something but rather, just the bare wire itself attaches to the back side of the patch panel, so then you just see a physical port on the front side into which you plug a regular cable. 

Again, this is used to make those actual connections in the back. 

< [Table of Contents](#top) | [References](#references) >

###### Optical Time Domain Reflectometer
***

An ``OTDR`` is used to characterize and find faults in fiber cabling.

It sends and measures signals, and you can actually detect where the damage is by the fact that it can measure the time it takes to reflect back, so that gives you a bit of an idea as to where any problem might be along the run. 

< [Table of Contents](#top) | [References](#references) >
<a id="LightMeter"></a>
***
###### Light Meter
***

A light meter is also used for installing and testing fiber optic networks but more so in terms of signal loss and the transmission capabilities in terms of i. the power required and ii. the continuity of the cable, itself. 

< [Table of Contents](#top) | [References](#references) >
<a id="ToneGenerator"></a>
***
###### Tone Generator
***

A tone generator and probe is used for network testing - particularly useful if you find yourself in an environment where they have not mapped where any given cable runs terminate.

A tone generator is used to determine where a cable is going, so, in other words, you see that there are several data ports in one room but you don't know where the other end is - might, of course, be in a wiring closet, but you don't physically know which ports are which, so this sends a strong magnetic signal down the cable, and then you can literally pick up on that on the other end. 

Can find where the other end of that cable is and you can also use this to find any cable interruptions, if present because, of course, the signal will not be detected on the other end, so you know that cable would be broken.

Can discover the transmission capability and the receiver sensitivity of those devices. 

The transmission capability and receiver sensitivity of devices are determined through equipment testing.

Resistance measurement is a fairly common when troubleshooting connectivity issues because if any two points of a connection don't have the same ohmage then you are likely to encounter problems, for sure. 

Again, useful when you just need to ascertain that electrical information.

In essence, this provides insight into the radio frequency performance of any given area, particularly with respect to how far the signal is carrying and/or how much it is being interfered with by anything that can cause electromagnetic interference, so you can get a bit of an idea as to what your implementation will look like and how it will perform by using the spectrum analyzer.

***
###### Port Scanner
***

After completing this topic, you should be able to: ``use a port scanner to identify which ports are open on a network``.

Port scanning is a technique used to identify which ports are open on a network because, of course, ports are used by a computer to send and receive information. 

So from the perspective of an attacker, a port represents a possible means of access and an analogy is, quite simply, knocking on each door in the neighborhood to see if anyone is home and if you are looking to intrude, then you won't intrude on a house where people are home but will focus your attention on one where no one is home. 

< [Table of Contents](#top) | [References](#references) >
<a id="AttacksEthicalHacking"></a>
***
###### Attacks - Ethical | Hacking
***

In terms of attacks, there are two types of approaches: 

i. ethical hacking, whereby an organization will perform port scanning to identify possible openings in their network but by doing so, they are trying to identify the vulnerabilities to correct them, so they can secure the network from these threats - by identifying possible openings in their network, organizations can implement changes to secure the network from outside threats, so they use the same techniques as hackers, but for a legitimate purpose. 

Hacking, of course, refers to identifying these potential openings for personal gain, so that you can exploit and access the network for unethical reasons. 

< [Table of Contents](#top) | [References](#references) >
<a id="VanillaTechnique"></a>
***
###### Scan - Vanilla Technique (Synchronization Flag - SYN - Initiate a Full Connection)
***

There is a method known as the ``Vanilla Technique``, which is a basic port scan, whereby a system attempts to connect to all ``65,536`` ports at once. 

This issues a full connect scan that sends requests to connect, in other words, it uses what's known as a ``SYN`` - ``Synchronization Flag`` - to actually initiate a full connection. 

These are fairly effective in terms of discovering which ports are open but they're also easily detectable because firewalls will always log a full connect request. 

< [Table of Contents](#top) | [References](#references) >
<a id="SYNScan"></a>
***
###### Scan - Synchronization SYN (Initiate a half-open Connection)
***

There's also a ``SYN`` scan, which is a half open scan - the attacking system will send a ``SYN`` request and wait for an acknowledgement response - the attacking system waits for a return ``SYN-ACK`` response. 

The scanner or the attacker does not respond to any of the received responses and as such, the system does not log the interaction, so in this way, the sender learns if a port is open or not without it being logged by that system. 

< [Table of Contents](#top) | [References](#references) >
<a id="ScanFIN"></a>
***
###### Scan - FIN - XMAS (Christmas)
***

There are also types of scans known as ``FIN`` scans and ``XMAS`` (or Christmas) scans and these ones are usually targeted on a specific port, but they use a variety of the flags that they use to initiate the connection.

An unsolicited ``FIN`` flag is sent to a system, whereby it can determine the state of the port based on the response - it's not trying to find every port, it's now trying to find a specific port and trying to obtain the state and the protocol information for that port - revealing the state of the port provides insight about the firewall.

A ``Christmas`` scan is simply sending a set of flags, usually the complete set of all flags - this ends up creating, in essence, a fairly nonsensical interaction, if you will - the system doesn't really know what to do, because it's receiving every possible type of connection request, so, again, determinations can be made based on the responses that come back. 

< [Table of Contents](#top) | [References](#references) >
<a id="ScanFTP"></a>
***
###### Scan - FTP Bounce
***

An ``FTP`` bounce scan allows the attacker to effectively disguise their location - sending the packets through an open ``FTP`` server so that the sender can perform their attack undetected. 

< [Table of Contents](#top) | [References](#references) >
<a id="ScanSweep"></a>
***
###### Scan - Sweep
***

A ``sweep`` scan sends a number of ``ping`` commands to the same port on multiple computers to identify which devices on the network are active - those that respond have an open port, so that presents the avenue of attack. 

< [Table of Contents](#top) | [References](#references) >
<a id="harden"></a>
***
###### Conclusion - Systems Hardened
***

Ideally, to reduce the chance of falling victim to port scanning attacks, you should really just ensure that your systems are hardened, in other words, if you do not use that port then, quite simply, disable it - for those that are open, ensure that they are configured correctly on your firewalls so that only the appropriate traffic is allowed through.

<a id="loopback"></a>
***
###### PING - Loopback
***

Can also ``ping`` yourself, so to speak, the loopback address is used to determine if ``TCP/IP`` is installed.

This is the same for every system running Windows - that address is reserved and, in essence, is bound to every adapter and it indicates if ``TCP/IP`` has correctly configured itself on the network interface. 

So if you are having connectivity issues, one of the first things to try is simply ``ping`` the loopback address and the syntax is, quite literally:


``ping 127.0.0.1`` 


Can, in many cases, also ``ping localhost`` - resolves to ``127.0.0.1`` and again, it just tells you that your own network interface has ``TCP/IP`` installed correctly on itself.


Then you can work out from that point and ``ping`` other IP addresses, or in fact, you can also, in most cases, ``ping`` a ``DNS`` name, so, for example, you could:


``ping server1.mydomain.com``


Even Internet names as well, you can:


``ping www.yourfavoritewebsite.com``


In many cases, Internet pinging might not yield any results because many routers on the Internet will simply discard ping packets to prevent them from being pinged too often.

###### ``ipconfig``
***

``ipconfig`` is perhaps one of the most common tools - used to determine your network configuration - displays the current TCP/IP address information and some of the main commands include:


- ``ipconfig /all``


- ``ipconfig /release``


- ``ipconfig /renew``


``ipconfig /all`` - just gives you much more detailed information - your IP address, your subnet mask and your default gateway but ``/all`` will give you your ``DNS`` server, your ``DHCP`` server, lease information with respect to when you got the address and when it's validity period expires, things along those lines.

Specifically, the ``/release`` and the ``/renew`` pertain to your ``DHCP`` configuration - if you feel like you received an incorrect or invalid address, you can discard the address by issuing: ``ipconfig /release`` then ``ipconfig /renew`` initiates a request for a new address from the ``DHCP`` server.

***
###### Traceroute (``tracert``)
***

``tracert`` or more commonly referred to as ``traceroute`` (linux), is a command that can show several details about the path that packets take, in other words, from your device to the destination.

This is a routing tool and again, often referred to as ``traceroute`` but it is, quite literally, the path of routers that are being crossed as a packet travels from the source to the destination.

This is generally only useful when you are troubleshooting network paths within your own internetworking environment, for example, if you feel like a package should travel this route and it's going all over the place, then you can possibly try to troubleshoot your routers to redirect it a little bit better. 

Can see the path that it takes over the Internet but, of course, there's not much you can do about changing the path on the Internet, but useful, again, for internal routing. 

< [Table of Contents](#top) | [References](#references) >

###### ``pathping`` (Multiple Echo Request messages)
***

``pathping`` is kind of similar to ``ping`` and ``tracert`` put together - its used to identify network latency and network loss issues. 

``Multiple Echo Request`` messages are sent to each router over a period of time and the results are computed based on information that is received back from those routers to identify problems with the router configuration or the way that you have implemented your subnets.

The command supports multiple parameters to refine it, but, in essence, it kind of combines ``ping`` with ``traceroute`` so that you can get statistical information about the route that packets take. 

< [Table of Contents](#top) | [References](#references) >

###### ``nslookup``
***

The ``nslookup`` command line tool is a network utility program.

``nslookup`` is used to diagnose ``DNS`` problems - ``Domain Name Service`` - returning the IP address that matches the hostname that you supply or vice versa but you can also verify which ``DNS`` server is responding to the request, so if you suspect that perhaps a rogue ``DNS`` server has been introduced into the environment, ``nslookup`` will tell you which server satisfied that request. 

This is only available, of course, if ``TCP/IP`` is installed, but it can also do reverse name lookup - you can return the IP address matching a name or you can return the name matching an IP address - either way, you still get the correct resolution of this address to that name. 

< [Table of Contents](#top) | [References](#references) >
<a id="arp"></a>
***
###### Address Resolution Protocol (ARP) - Resolves the IP address to a MAC address
***

``arp`` refers to the ``Address Resolution Protocol`` - this is a command used to manage our cache information for network devices - the ``arp`` command line tool is an external command.

Can display the cache, or you can in fact add or remove a value from the cache - recall that ``ARP resolves the IP address to a MAC address``.

It's possible that a system might have changed its IP address, but yet the cache still believes that it's using its previous IP address, so if that's the case, then packets will be routed to the incorrect address, so you can purge out the cache entirely, or you could just remove that single entry. 

Conversely, if you have a system that you know always has the same IP address and it's always attached to the same MAC address, then you can actually add persistent entries to the cache and they'll be resolved more quickly. 

< [Table of Contents](#top) | [References](#references) >

``netstat`` is used to display and check if various aspects of ``TCP/IP`` are working, such as your network connections, both incoming and outgoing - can see the status of your routing tables - can also see the statistics of your network interface with respect to the protocols and the port status - can see which ports are open and which protocols are being used over those ports.

<a id="route"></a>
***
###### ``route``
***

``route`` is used to manually configure a routing table - this is typically only used when you have a Windows Server, for example, that is being used as a router, which isn't particularly common but can certainly still be done, so this is used to quite literally ``add``, ``delete``, or ``change`` the static routing entries that you have for that server, for example, you can issue the command ``route add`` to manually add a new route. 


- ``<destination>`` is the network to which this packet is destined


- ``mask`` is the subnet mask in use for that destination network


- ``<gateway>`` is, in essence, the default gateway that you see from your perspective to get out


- ``metric`` is the distance how many hops


- ``<interface>`` is the one that you see as being the doorway into that network


```
route add <destination> mask <netmask> <gateway> metric <interface>
```

So you quite literary specify those values, that adds the route, and that server now knows how to route a packet to that destination network.
